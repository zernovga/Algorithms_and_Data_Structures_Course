## 4. Найти сложность приведенной ниже программы:

```python
def funct(n):
    count = 0
    for i in range(n//2, n+1): # O(n)
        j = 1
        while j <= n: # O(log(n))
            j = 2 * j
            k = 1
            while k <= n: # O(log(n)) => O(n*log^2(n))
                k = 2 * k
                count += 1
```
